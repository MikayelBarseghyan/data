```python
import sqlite3
import pandas as pd
```

```python
conn = sqlite3.connect("tips.db")
cur = conn.cursor()
```

```python
query="SELECT name FROM sqlite_master WHERE type='table';"
cur.execute(query)
print(cur.fetchall())
```

```python
query = """SELECT * FROM Observation WHERE Observation.total_bill > 50;"""
pd.read_sql(query, conn)
```

```python
query = """SELECT AVG(Observation.tip) as Mijin FROM Observation;"""
pd.read_sql(query, conn)
```

```python
query = """SELECT MAX(Observation.total_bill) as Maximum_bill FROM Observation;"""
pd.read_sql(query, conn)
```

```python
query = """SELECT SUM(Observation.size) as people, Day.day FROM Observation
JOIN Day ON Day.day_id=Observation.day_id
GROUP BY Day.day;"""
pd.read_sql(query, conn)
```

```python
query = """SELECT AVG(Observation.total_bill) AS average_bill, Observation.size FROM Observation
GROUP BY  Observation.size;"""
pd.read_sql(query, conn)
```

```python
query = """SELECT SUM(Observation.tip) as Tip, Smoker.smoker FROM Observation
JOIN Smoker ON Smoker.smoker_id=Observation.smoker_id
GROUP BY Smoker.smoker;"""
pd.read_sql(query, conn)
```

```python
query = """SELECT (Observation.tip)/(Observation.total_bill)*100 as Averageperc_tip, Day.day FROM Observation
JOIN Day ON Day.day_id=Observation.day_id
GROUP BY  Day.day;"""
pd.read_sql(query, conn)
```

# **<font color='red'> For average calculation you should either divide the sum of tip to the sum of total bill or use avg aggregate function on the division of tip to total bill. Minus 1 point </font>**

```python
query = """SELECT Observation.total_bill, Day.day, Time.time
FROM Observation
JOIN Day ON Observation.day_id = Day.day_id
JOIN Time ON Observation.time_id = Time.time_id;"""
pd.read_sql(query, conn)
```

```python
query = """SELECT Sex.sex, COUNT(Observation.sex_id) AS Total_observations
FROM Observation
JOIN Sex ON Observation.sex_id = Sex.sex_id
GROUP BY Sex.sex;"""
pd.read_sql(query, conn)
```

```python
query = """SELECT Observation.tip, Day.day, Time.time
FROM Observation
JOIN Day ON Observation.day_id = Day.day_id
JOIN Time ON Observation.time_id = Time.time_id
WHERE Day.day = 'Sat' AND Time.time = 'Dinner';"""
pd.read_sql(query, conn)
```

```python
query = """SELECT Sex.sex, MAX(Observation.total_bill) AS largest_total_bill
FROM Observation
JOIN Sex ON Observation.sex_id = Sex.sex_id
GROUP BY Sex.sex;"""
pd.read_sql(query, conn)
```

```python
query = """SELECT AVG(Observation.tip) AS tip_avg
FROM Observation
WHERE Observation.size > 4;"""
pd.read_sql(query, conn)
```

```python
pd.read_sql_query('''SELECT Observation.* , day, time FROM Observation
JOIN Day ON Observation.day_id = Day.day_id
JOIN Time ON Observation.time_id = Time.time_id
WHERE Day.day IN ('Fri', 'Sat')
AND Time.time IN ('Lunch', 'Dinner')''', conn)
```

```python
query = """SELECT * FROM Observation
WHERE (Observation.tip / Observation.total_bill) > 0.20;"""
pd.read_sql(query, conn)
```
