```python
mtqer = []
for page_num in range(1,11):
    page_url =  base_url + str(page_num)
    response = requests.get(page_url)
    soup = BeautifulSoup(response.content, "html.parser")

    quotes = soup.find_all("div", class_="quote")
    for mitq in quotes:
        mtqer_text = mitq.find("span", class_="text").get_text()
        author = mitq.find("small", class_="author").get_text()
        tags = [tag.get_text() for tag in mitq.find_all("a", class_="tag")]

        author_link = "https://quotes.toscrape.com" + mitq.find("a")["href"]
        author_response = requests.get(author_link)
        author_soup = BeautifulSoup(author_response.content, "html.parser")

        born_date = author_soup.find("span", class_="author-born-date").get_text()
        born_location = author_soup.find("span", class_="author-born-location").get_text()
        description = author_soup.find("div", class_="author-description").get_text().strip()

        mtqer.append({
            "Quote": mtqer_text,
            "Author": author,
            "Tags": tags,
            "Born": born_date,
            "location" : born_location,
            "Description": description
        })
```

```python
df = pd.DataFrame(mtqer)
print(len(df))
df
```
